```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.datasets import imdb
from tensorflow.keras.models import load_model
```

```python
#loading imdb datasets
word_index= imdb.get_word_index()
reverse_word_index= {value:key for key, value in word_index.items()}
```

```python
##Loading pre-trained model using Relu Activation
model= load_model('simple_rnn.h5')
model.summary()
```

```python
model.get_weights()
```

```python
#Function to decode reviews
def decode_review(encoded_review):
    return ' '.join(reverse_word_index.get(i - 3, '?') for i in encoded_review)
```

```python
#function to preprocess user input
def preprocess_input(text):
    words= text.lower().split()
    encoded_review= [word_index.get(word, 2) + 3 for word in words]
    padded_review= sequence.pad_sequences([encoded_review], maxlen=500)
    return padded_review
```

```python
# prediction function
def prediction(text):
    preprocess_text=preprocess_input(text)
    predicted_val= model.predict(preprocess_text)
    sentiment= 'Positive' if predicted_val[0][0]>0.5 else 'Negative'
    return sentiment, predicted_val[0][0]
```

```python
# Step 4: User Input and Prediction
# Example review for prediction
example_review = "this film was absolutely wonderful and i loved every moment of it"

sentiment,score=prediction(example_review)

print(f'Review: {example_review}')
print(f'Sentiment: {sentiment}')
print(f'Prediction Score: {score}')
```
